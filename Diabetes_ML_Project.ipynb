```python
inputs = df.drop("Outcome",axis = 1) 
#Define input features
```

```python
y = df['Outcome']
#Defines pur expected answer
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,y,random_state=42,test_size=.2)
```

```python
import tensorflow as tf
```

```python
inputs = tf.keras.Input(shape=(8))
# redefined 'inputs' we use this function to initialize to reate an input layer
x = tf.keras.layers.Dense(1000,activation='relu')(inputs)
x = tf.keras.layers.Dense(1000,activation='relu')(x)
x = tf.keras.layers.Dense(1000,activation='relu')(inputs)
x = tf.keras.layers.Dense(1000,activation='relu')(inputs)
x = tf.keras.layers.Dense(1000,activation='relu')(inputs)

#we define x using a Dense Layer with 1000x1000 neurons
#Actvation as 'relu'
output = tf.keras.layers.Dense(1,activation='sigmoid')(x)
#define output with a layer in 'sigmoid' for binary prediction
```

```python
model = tf.keras.Model(inputs,output)
#We define model as its input layer and output layer
```

```python

model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=["accuracy",tf.keras.metrics.BinaryAccuracy()]
              )
```

```python
attempt_1 = model.fit(
    X_train,y_train,
    epochs=10000,
    validation_data=(X_test,y_test),
    callbacks=[tf.keras.callbacks.EarlyStopping(patience=50)])
```
