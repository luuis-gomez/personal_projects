```python
import sklearn
import pandas as pd
```

```python

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
hp = pd.read_csv(r"/Users/luisgomez/Documents/house_pricing/housing.csv", header = None, delimiter = r"\s+", names = column_names)
```

```python
hp
```

```python
X = hp.drop(["PRICE"],axis=1)
```

```python
#Definimos y para entrenarlo
y = hp['PRICE']
y
```

```python
#importamos de sklearn el learn split
from sklearn.model_selection import train_test_split
```

```python
#Esta linea de comando aleatoriza datos y los divide en datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
clf = Pipeline(
    [
      
        ("clf", LinearRegression()),
    ]
)
```

```python
clf.fit(X_train,y_train) #X_train son las caracteristicas y y_train son las respuestas
```

```python
clf.score(X_test,y_test)
```
